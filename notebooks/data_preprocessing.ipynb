```python
import pandas as pd


dga = pd.read_csv("/content/dga_data.csv")

# Keep only required columns
dga = dga[['domain', 'subclass']]

# Remove Alexa entries (these are legit domains mistakenly inside DGA file)
dga = dga[~dga['subclass'].isin(['alexa', 'legit'])]

dga['label'] = 1

print("DGA shape after removing alexa:", dga.shape)
print("\nDGA family distribution:")
print(dga['subclass'].value_counts())


legit = pd.read_csv("/content/tranco_JLK5Y.csv")

# Rename columns properly
legit.columns = ['index', 'domain']

# Keep only domain column
legit = legit[['domain']]

# Sample legit domains equal to DGA count
legit = legit.sample(n=len(dga), random_state=42)

# Assign legit labels
legit['subclass'] = 'legit'
legit['label'] = 0

print("\nLegit shape after sampling:", legit.shape)


final_df = pd.concat([dga, legit], ignore_index=True)

# Shuffle dataset
final_df = final_df.sample(frac=1, random_state=42).reset_index(drop=True)

print("\nFinal Dataset Shape:", final_df.shape)
print("\nLabel Distribution:")
print(final_df['label'].value_counts())

print("\nSubclass Distribution:")
print(final_df['subclass'].value_counts())

final_df.to_csv("final_dga_dataset_cleaned.csv", index=False)
```
